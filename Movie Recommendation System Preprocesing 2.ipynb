```python
new_df['cast']=new_df['cast'].apply(lambda x: eval(x))

new_df['crew']=new_df['crew'].apply(lambda x: eval(x))


new_df['genres']=new_df['genres'].apply(lambda x: eval(x))


```

```python
def evaluate_genres(x):
    gem=[]
    sep=" "
    for i in x:
        if (i.get("name")=="Science Fiction"):
            
            gem.append("Sci-Fi")
        else:
            gem.append(i.get("name"))
            
    if (gem==[]):
        return np.NaN
    else:
            
        return (sep.join(gem))
    
```

```python
new_df['genres']=new_df['genres'].apply(lambda x:evaluate_genres(x))
```

```python
new_df['genres']
```

```python
def actor(x,n):
    cast=[]
    sep=" "
    for i in x:
        cast.append(i.get("name"))
    if(n==1):        
        if (cast==[]):
            return np.NaN
        else:
            return (cast[0])
    elif(n==2):
        if (len(x)<=1):
            return np.NaN
        else:
            return (cast[1])
    else:    
        if (len(x)<=2):
            return np.NaN
        else:
            return (cast[2])
```

```python
new_df['actor_1_name']=new_df["cast"].apply(lambda x:actor(x,1))
```

```python

new_df['actor_2_name']=new_df["cast"].apply(lambda x:actor(x,2))
new_df['actor_3_name']=new_df["cast"].apply(lambda x:actor(x,3))
```

```python
new_df
```

```python
def director(x):
    sep=" "
    dirr=[]
    for i in x:
        if(i.get("job")=="Director"):
            dirr.append(i.get("name"))
            
    if  (dirr==[]) :
        return np.NaN
    else:
        return (sep.join(dirr))
          
```

```python
new_df["director_name"]=new_df["crew"].apply(lambda x:director(x))
```

```python
new_df=new_df[['genres','title','actor_1_name','actor_2_name','actor_3_name','director_name']]
```
